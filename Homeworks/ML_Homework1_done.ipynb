3) What are the test and validation set, and why would you want to use them?  

**Test seti; train seti ile eğitilmiş olan makine öğrenmesi modelinin performansının sınanması için kullanılan, train seti ile aynı olasılık dağılımını taşıyan ve train setinden bağımsız olan veri setidir. Validation set ise; train seti ile aynı olasılık dağılımını taşıyan, overfitting durumunu engelleyebilmek için ve hiperparametre optimizasyonunu sağlayabilmek için kullanılan veri setidir.**  

4) What are the main preprocessing steps? Explain them in detail. Why we need to prepare our data?  

**1.Duplicate olmuş gözlemler incelenir, varsa silinir. Aksi taktirde modelde yanlılığa sebep olabilirler.**  

**2.Veri setindeki hedef değişkendeki değerlerin dağılımlarına ve frekanslarına bakılır. Eğer bu dağılım dengesizse bu duruma müdahale edilir. Gerektiğinde oversampling(smote) ya da undersampling metotları kullanılarak hedef değişken dengeli hale getirilmeye çalışılır.**  

**3.Eksik verilere müdahale edilir. Öncelikle verinin eksik olmasının bir başka değişkenle ilişkili olup olmadığı(rassal) incelenir. Rassal sebeplerle ortaya çıkan eksik veriler medyan değeri(numerik verilerde) veya mode(kategorik verilerde) değeri ile doldurulabilir. Rassal olmayan sebeplerle ortaya çıkan eksik verilerde ise söz konusu değişkende eksik olan gözlemler ve eksik olmayan gözlemler iki ayrı sınıfmış gibi değerlendirilip numaralandırılarak yeni bir değişken oluşturulabilir. Bununla birlikte eksik verileri doldurabilmek için yine makine öğrenmesi algoritmalarından yararlanılabilir. Veri setinin çok büyük boyutta olduğu durumlarda az sayıdaki eksik verinin silinerek veriden atılması da bir başka seçenek olarak değerlendirilebilir.**  

**4.Aykırı verilere müdahale edilir. Bazı algoritmalar aykırı verilere karşı hassastır ve bu durum modelin başarısını olumsuz yönde etkileyebileceği için aykırı verilerin veri setinden atılması uygun olacaktır. Aykırı verileri tespit etmede ve onları veri setinden atmada en sık kullanılan yöntemler 3-sigma, IQR metodu ve IsolationForest algoritması olarak gösterilebilir. 3-sigma yöntemi normal dağılım gösteren verilerde ortalamaya standart sapmanın 3 katının eklenmesi ve çıkarılmasıyla bulunan değerlerin bir sınır olarak kabul edilip bu sınırın dışında kalanların aykırı veri olarak tanımlanıp veri setinden atılmasına dayanmaktadır. IQR metodunda ise; verideki 1. ve 3. kuantil değerlerinin farkının 1.5 ile çarpılmasıyla elde edilen değerin  1.kuantil değerinden çıkarılması ve 3.kuantil değerine eklenmesiyle bulunan yeni değerlerin sınır olarak kabul edilmesine ve bu sınırların dışında kalan gözlemlerin aykırı veri olarak nitelendirilip veri setinden atılmasına dayanmaktadır. IsolationForest ise karar ağacı temelli bir algoritma olup gözlemlerin aykırı olup olmadığını sınıflandıracaktır. Böylece IsolationForest'ın aykırı gözlem olarak tespit ettiği(yani algoritmanın -1 olarak etiketlediği) veriler istenildiğinde veri setinden atılacaktır.**  

**5.Feature Scaling. Scaling işlemleri değişkenlerin çok farklı değer aralıklarında olması, model başarısının arttırılması ve modelin eğitim süresinin hızlandırılması gibi nedenlerle uygulanmaktadır. Scaling başlığı altında Standardizasyon ve Normalizasyon yöntemlerinden bahsedebiliriz. Standardizasyon yöntemi, normal dağılım gösteren verilere uygulanarak onları ortalaması 0 ve standart sapması 1 olan bir biçime dönüştürür. Normalizasyon yöntemi, genelde normal dağılım göstermeyen verilerde değerlerin 0-1 aralığına sıkıştırılması için kullanılmaktadır. Bu yöntemler nümerik değişkenlerde uygulanırlar.**  

**6.Feature Encoding. Makine öğrenmesi modelleri oluştururken metin formundaki kategorik verilerin numaralarla etiketlendirilmesi gerekmektedir, aksi taktirde modeli kuramayız. Bu numaralandırmaları yaparken rastgele bir şekilde değil belli başlı bazı kurallar doğrultusunda yapmamız icap etmektedir. Kategorik veriler kendi aralarında nominal ve ordinal olarak iki gruba ayrılmaktadırlar. Nominal verilere örnek olarak cinsiyet, renk, taraftarı olunan futbol takımı verilebilir. Nominal değerlerin numaralandırılmasında kullanılan tekniğe One Hot Encoding denilir. OneHotEncoding yönteminde söz konusu kategorik değişkendeki her bir unique sınıf için yeni bir kolon oluşturulur ve bu kolonlardan bir tanesi silinir. Bunun nedeni geri kalan kolonların silinen kolonu da açıklayabilmelerinden kaynaklanmaktadır.OneHotEncoding yönteminde belli bir gözlemdeki söz konusu kategorik sınıf kendisi için oluşturulan kolonda 1 olarak numaralandırılı ve diğer kolonlarda da 0 olarak numaralandırılma yapılır.OneHotEncoding için oluşturulan kolonlardaki değerlerin hepsi 0 olduğunda o gözlemdeki kategorik sınıfın en başta kolonunu sildiğimiz kategorik sınıf olduğu anlaşılır. Ordinal verilerde ise bir üstünlük sıralaması mevcuttur. Ordulardaki askeri rütbeler bu kategorik veri tipine örnek olarak gösterilebilir. Burada kategorik sınıflar en alt rütbeden en üst rütbeye doğru sıralandırılarak sıralı bir şekilde numaralandırılır. Bununla birlikte hedef değişkenin kategorik bir veri olması halinde o değişkenin nominal veya ordinal olmasına bakılmaksızın her bir sınıfa farklı bir numara gelecek şekilde sıralı olarak numaralandırılma yapılır.**  

**7.Train, Test ve Validation verilerinin ayrıştırılması. Makine Öğrenmesi modelleri oluşturulurken elimizdeki veri setinin bir kısmıyla modeli eğitip bir kısmıyla da modelin başarısını sınamamız gerekmektedir. Train seti modelin eğitilmesi için kullanılırken test seti ise, modelin başarısının sınanmasında kullanılmaktadır. Bununla birlikte overfitting durumunu önlemek, veri sızıntızı durumuna engel olmak ve hiperparametre optimizasyonunu daha doğru şekilde yapabilmek için Validation seti denilen ayrı bir alt veri kümesini ana veri setinden ayırmakta yarar vardır. Bahsi geçen 3 ayrı veri kümesinde de olasılık dağılımı aynı olmalıdır.Bu veri kümelerini ayırırken Train setinin ana veri setinin %50'sinden daha azı olmaması gerektiği unutulmamalıdır. En uygun ayırma miktarı Train: %60, Test: %20, Validation: %20 olarak düşünülebilir.**  

**Peki biz bu kadar işlemi neden uyguluyoruz? Preprocessing işlemleri veri setinin daha anlamlı/açıklanabilir hale gelmesi, veri setindeki hatalı kısımların onarılması, veri setinin modelleme aşamasında kullanılacak olan algoritmanın varsayımlarını sağlaması, oluşturulacak makine öğrenmesi modelinin başarısının optimum seviyede olması ve makine öğrenmesi modelini oluşturduğumuz programlama dilinin ve modüllerin gerektirdiği şartların karşılanması amacıyla uygulanmaktadır. Unutulmamalı ki, preprocessing işlemleri bir makine öğrenmesi projesinde en çok vakit harcanan süreç olup, modelin başarısını kritik şekilde etkileyebilecek müdahaleleri içermektedir.**
